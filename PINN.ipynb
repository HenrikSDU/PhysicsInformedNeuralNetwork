```python
from scipy.constants import hbar
def psi_analytical_n(x,n=0,m=1.0,w=1):
    y = np.sqrt(m*w/1)*x
    if n == 0:
        return np.exp(-0.5*y**2)
    if n == 1:
        return 2*y*np.exp(-0.5*y**2)
    if n == 2:
        return (4*y**2-2)*np.exp(-0.5*y**2)
   
a = 5
x = np.linspace(-a,a,199)
psi = psi_analytical_n(x,n=0)
plt.plot(x,psi)
big = np.sqrt(1/hbar) 
```
